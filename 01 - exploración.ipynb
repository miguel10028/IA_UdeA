```python
#Importamos librerias
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Lectura del archivo "train.csv"

nombre_archivo = "train.csv"

#Creamos el DataFrame
data = pd.read_csv(nombre_archivo)

# Mostrar un breve resumen del archivo "train.csv"
data
```

# **Exploración data archivo "train.csv":**


```python
#Consultamos columnas en el archivo "train.csv":
print("Columnas del archivo", nombre_archivo,":\n", data.columns)

#Información dimensiones archivo "train.csv":
print("\nTamaño del archivo:",
      "\nFilas:", data.shape[0],
      "\nColumnas:", data.shape[1])

#Verificar datos faltantes
print("\nDatos Faltantes por columna:\n",data.isnull().sum())
```

```python
# Distribución de los programas académicos
program_dist = data['ESTU_PRGM_ACADEMICO'].value_counts()
print("Distribución de los estudiantes en los diferentes programas académicos:\n", program_dist)
```

```python
# Distribución de los estratos socioeconómicos
estrato_dist = data['FAMI_ESTRATOVIVIENDA'].value_counts()
print("\nDistribución de los estratos socioeconómicos:\n", estrato_dist)

# Histograma de estratos
plt.figure(figsize=(5, 4))
sns.countplot(x='FAMI_ESTRATOVIVIENDA', data=data)
plt.title('Distribución de Estratos Socioeconómicos')
plt.xticks(rotation=90)
plt.show()
```

```python
# Boxplot de horas de trabajo vs valor de la matrícula
plt.figure(figsize=(10, 5))
sns.boxplot(x='ESTU_VALORMATRICULAUNIVERSIDAD', y='ESTU_HORASSEMANATRABAJA', data=data)
plt.xticks(rotation=90)
plt.title('Horas de Trabajo Semanal vs Valor de la Matrícula')
plt.show()
```
